

This notebook just generates a plot to visualize the timing of events in this piece.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This implementation is very unclean.
# Could clean it up if we actually end up using this piece.

NUM_BARS = 8
TIME_SIG_TOP = 4
TIME_SIG_BOT = 4

TIME_DIVS_TOTAL = NUM_BARS * TIME_SIG_TOP

BPM = 120
BPS = BPM / 60 # beats per second

tt = range(TIME_DIVS_TOTAL)
tt_s = np.array(tt) / BPS

MEASURE = [0]*TIME_SIG_TOP
MEASURE[0] = 1
ALL_MEASURES = MEASURE*NUM_BARS
ALL_MEASURES_s = np.array(ALL_MEASURES) / BPS

plt.figure(figsize=(8, 4))
plt.stem(tt_s, ALL_MEASURES_s * 0.1, 'k', label='beat') #markerfmt='k.')


SNARE_HITS = [] #range(NUM_BARS)
offset = 0
for bar_ndx in range(NUM_BARS):
    SNARE_HITS.append(bar_ndx*TIME_SIG_TOP + offset)
    offset += 1/2 #/TIME_SIG_BOT/2
SNARE_HITS_s = np.array(SNARE_HITS) * 60 / BPM
R_HITS_s = SNARE_HITS_s[0::2]
L_HITS_s = SNARE_HITS_s[1::2]

plt.stem(R_HITS_s, np.ones_like(R_HITS_s), 'r', label='Right Hand')
plt.stem(L_HITS_s, np.ones_like(L_HITS_s), label='Left Hand') #, 'r')
plt.xlabel('Time [s]')
plt.ylabel('Event')
plt.tick_params(
    axis='y',          
    which='both',      
    left=False,      
    labelleft=False) 

plt.legend()
plt.show()
```
